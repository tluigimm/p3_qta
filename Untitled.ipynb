```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_excel("tabelas/RelatorioQualidadeAguasSuperficiais 2001 2005.xlsx")
df2 = pd.read_excel("tabelas/RelatorioQualidadeAguasSuperficiais 2006 - 2010.xlsx")
df3 = pd.read_excel("tabelas/RelatorioQualidadeAguasSuperficiais 2011 - 2015.xlsx")
df4 = pd.read_excel("tabelas/RelatorioQualidadeAguasSuperficiais 2016 2020.xlsx")
```

```python
df_list = [df1, df2, df3, df4]
dfinal = pd.concat(df_list)
```

```python
dfinal['Data Coleta'] = pd.to_datetime(dfinal['Data Coleta'])
dfinal["Data Coleta"] = dfinal["Data Coleta"].astype('datetime64[ns]')
dfinal.sort_values(by="Data Coleta")

dfinal.Valor = dfinal.Valor.replace({',':'.'}, regex=True)

dfinal.head()
```
